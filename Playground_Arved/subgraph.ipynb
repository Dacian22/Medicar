```python
from submap import Submap
```

```python
# define parameters for the subgraph
subgraph_params = {
                'special_nodes': ['Zwischen den Räumen', 'Stimme vom Berg',
                                'Emmaus Kapelle', 'Klinik für Frauenheilkunde [1]',
                                'Ernst von Bergmann'],
                'allowed_highway_types': ['footway', 'service']
}

# define parameters for the shortest path
shortest_path_params = {
                'start_node': 'Zwischen den Räumen',
                'end_node': 'Ernst von Bergmann'
}
```

```python
# create the submap
submap = Submap(**subgraph_params)
# determine closest nodes to the special nodes
closest_nodes = submap.closest_node_to_special_nodes()
# create the subgraph
subgraph = submap.create_graph()
# reduce the subgraph to the largest connected component
subgraph = submap.keep_largest_connected_component(subgraph)
# label edges of the graph
edge_labels = submap.label_edges(subgraph)
# calculate the shortest path
shortest_path = submap.find_astar_path(subgraph, **shortest_path_params)
```

```python
# plot the graph
submap.plot_graph(subgraph, edge_labels, shortest_path)
```

```python
submap.translate_path_to_mqtt(shortest_path)
```
